Processing the business dataset, to find US popular open restaurants.

```python
import pandas as pd
df_yelp_business = pd.read_json('yelp_academic_dataset_business.json', lines=True)
df_yelp_business
```

```python
# categories column has many categories at once. Let's split the categories column and store each category in it's own column
business_category_split = df_yelp_business['categories'].str.split(',', expand=True)[[0,1,2]]
# rename the new columns
business_category_split.columns = ['category_1', 'category_2', 'category_3']
df_yelp_business = pd.concat([df_yelp_business, business_category_split], axis=1)
# now drop the actual categories column
df_yelp_business = df_yelp_business.drop(['categories'], axis=1)
# print the head of the dataset
df_yelp_business.head()
```

```python
df_yelp_business['city']
```

```python
# let's filter our dataset.
# we'll use Restaurants category
df_yelp_restaurants = df_yelp_business.loc[(df_yelp_business['category_1']=='Restaurants') | (df_yelp_business['category_2']=='Restaurants') | (df_yelp_business['category_3']=='Restaurants')]
df_yelp_restaurants
```

```python
city_cnt = df_yelp_restaurants['city'].value_counts()
print(city_cnt)
state_cnt = df_yelp_restaurants['state'].value_counts()
print(state_cnt)
```

```python
open_cnt = df_yelp_restaurants['is_open'].value_counts()
print(open_cnt)
# us_restaurants = df_yelp_restaurants.drop(df_yelp_restaurants[df_yelp_restaurants['is_open'] == 0].index)
# us_restaurants
```

```python
print('Mean of restaurants review_count is: ', df_yelp_restaurants['review_count'].mean())
print('Median of restaurants review_count is: ', df_yelp_restaurants['review_count'].median())
print(df_yelp_restaurants['review_count'].value_counts())
```

```python
pop_rests = df_yelp_restaurants.drop(df_yelp_restaurants[df_yelp_restaurants['review_count'] <= 20].index)
pop_rests.to_csv("pop_rest.csv", index = None)
```

Process the review dataset, only keep reviews for our target restaurants.

```python
import json
review_file = open("yelp_academic_dataset_review.json", encoding='utf-8')
review = []
for line in review_file:
    review.append(json.loads(line))

df_yelp_reviews = pd.DataFrame(review)
review_file.close()
df_yelp_reviews
```

```python
pop_reviews = df_yelp_reviews[df_yelp_reviews['business_id'].isin(pop_rests['business_id'])]
pop_reviews
```

```python
pop_reviews.to_csv("pop_reviews.csv", index=None)
```

```python
import pandas as pd
reviews_df = pd.read_csv('pop_reviews.csv')
df_res_bus = pd.read_csv('pop_rest.csv')
df_merge = pd.merge(reviews_df, df_res_bus[['city', 'business_id',"review_count"]], how = "outer", on ="business_id")
df_merge
```
